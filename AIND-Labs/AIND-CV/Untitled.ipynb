# Creating a filter, Edge detection

## Import resources and image display

```python
import matplotlib.pyplot as plt
import numpy as np
import cv2

%matplotlib inline

image = cv2.imread('image/city_hall.jpg')

# Make a copy of the image
image_copy = np.copy(image)

# Change the color channel order
image_copy = cv2.cvtColor(image_copy, cv2.COLOR_BGR2RGB)

# Display the image
plt.imshow(image_copy)
```

```python
# Convert to gray scale for filter
gray = cv2.cvtColor(image_copy, cv2.COLOR_RGB2GRAY)

plt.imshow(gray, cmap='gray')
```

```python
# 3x3 filter for vertical edge detection
# Sobel filter
sobel_x = np.array([[-1, 0, 1],
                    [-2, 0, 2],
                    [-1, 0, 1]])

# Perform convolution on the image
filtered_image = cv2.filter2D(gray, -1, sobel_x)

plt.imshow(filtered_image, cmap='gray')
```

## Create binary image

```python
# Create a threshold that set all filtered values to white
retval, binary_image = cv2.threshold(filtered_image, 100, 250, cv2.THRESH_BINARY)

plt.imshow(binary_image, cmap='gray')
```
