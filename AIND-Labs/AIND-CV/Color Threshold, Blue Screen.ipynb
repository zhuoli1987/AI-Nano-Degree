# Color Threshold, Blue Screen

## Import Resources

```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

## Read in and display the image

```python
# Load the image
image = cv2.imread('image/pizza.jpg')

# Print the image info
print('This image is: ', type(image), ' with dimensions: ', image.shape)
```

```python
%matplotlib inline

# Make a copy of the image
image_copy = np.copy(image)

# Change the color channel order
image_copy = cv2.cvtColor(image_copy, cv2.COLOR_BGR2RGB)

# Display the image
plt.imshow(image_copy)
```

## Define the color threshold

```python
# Define the selection boundary in RGB value
lower_blue = np.array([0,0,10])
higher_blue = np.array([80,100,250])
```

## Create a mask

```python
# Define the mask
mask = cv2.inRange(image_copy, lower_blue, higher_blue)

# Visualize
plt.imshow(mask, cmap='gray')
```

```python
masked_image = np.copy(image_copy)

masked_image[mask != 0] = [0, 0, 0]

plt.imshow(masked_image)
```

## Mask and add a background

```python
background_image = cv2.imread('images/background.jpg')
background_image = cv2.cvtColor(background_image, cv2.COLOR_BGR2RGB)

crop_background = background_image[0:568, 0:800]

crop_background[mask == 0] = [0, 0, 0]
```

```python
complete_image = crop_background + masked_image

plt.imshow(complete_image)
```
